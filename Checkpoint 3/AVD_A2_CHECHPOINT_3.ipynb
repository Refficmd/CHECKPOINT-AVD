# 3. Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dikarenakan presentase yang tidak terlalu tinggi namun merupakan variable yang penting, maka dilakukan imputasi dengan metode yang paling sesuai, yaitu Winsorizing karena tetap mempertahankan struktur data tanpa membuang nilai ekstrem

# Melakukan imputasi

```python
columns_to_impute = ["Calories_Burned", "BMI"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

# Pengecekan ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# Relationship

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Session_Duration (hours)'], y=df['Calories_Burned'], alpha=0.6)
plt.xlabel("Session Duration (hours)")
plt.ylabel("Calories Burned")
plt.title("Relasi antara durasi Latihan dan Kalori yang Terbakar")
plt.show()
```

Scatter plot menunjukkan adanya korelasi positif antara durasi sesi latihan dan jumlah kalori yang terbakar, yang berarti semakin lama seseorang berlatih, semakin banyak kalori yang mereka bakar. Namun, meskipun hubungan ini kuat, terdapat variasi dalam jumlah kalori yang terbakar untuk durasi yang sama, yang bisa disebabkan oleh faktor lain seperti intensitas latihan, jenis latihan, atau kondisi fisik individu. Berdasarkan temuan ini, gym dapat memberikan rekomendasi personalisasi durasi dan intensitas latihan untuk membantu anggota mencapai target pembakaran kalori yang lebih optimal. Selain itu, edukasi mengenai efisiensi latihan, seperti bagaimana meningkatkan intensitas dalam waktu yang lebih singkat, dapat membantu anggota yang memiliki keterbatasan waktu tetap mencapai hasil yang maksimal.
